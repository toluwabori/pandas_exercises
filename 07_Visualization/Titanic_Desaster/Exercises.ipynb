# Visualizing the Titanic Disaster

### Introduction:

This exercise is based on the titanic Disaster dataset avaiable at [Kaggle](https://www.kaggle.com/c/titanic).  
To know more about the variables check [here](https://www.kaggle.com/c/titanic/data)

### Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv)

```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
url ='https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv'
```

### Assign it to a variable titanic 

```python
df_titanc = pd.read_csv(url)
```

### Set PassengerId as the index 

```python
df_titanc.set_index('PassengerId')
```

### Create a pie chart presenting the male/female proportion

```python
male =(df_titanc.Sex == 'male').sum()
femae = (df_titanc.Sex == 'female').sum()
proportions = [male,femae]
plt.pie(x=proportions,labels=['Male','Female'],)
plt.title('PROPORTYION OF MALE TO FEMALE');
```

### Create a scatterplot with the Fare payed and the Age, differ the plot color by gender

```python
c=sns.scatterplot(x=df_titanc.Age,y=df_titanc.Fare,hue=df_titanc.Sex)
c.set(title='fare x age');
```

### How many people survived?

```python
sns.countplot(x=df_titanc.Survived);
plt.title('HOW MANY SURVIVED')
```

### Create a histogram with the Fare payed

```python
sns.histplot(x=df_titanc.Fare)
```
