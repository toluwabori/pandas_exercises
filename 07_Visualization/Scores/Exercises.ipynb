### Create the DataFrame that should look like the one below.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'], 
            'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'], 
            'female': [0, 1, 1, 0, 1],
            'age': [42, 52, 36, 24, 73], 
            'preTestScore': [4, 24, 31, 2, 3],
            'postTestScore': [25, 94, 57, 62, 70]
          }
```

```python
df_data = pd.DataFrame(data)
```

```python
df_data
```

### Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

```python
plt.scatter (x=df_data.preTestScore,y=df_data.postTestScore,s=df_data.age)
plt.xlabel('posttestscore')
plt.ylabel('prtestscore')
```

### Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

```python
plt.scatter (x=df_data.preTestScore,y=df_data.postTestScore,s = df_data.postTestScore*4.5,c=df_data.female)
plt.xlabel('posttestscore')
plt.ylabel('prtestscore');
```
