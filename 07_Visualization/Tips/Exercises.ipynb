### Assign it to a variable called tips

```python
import pandas as pd
import seaborn as sns 
url ='https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv'
```

```python
import matplotlib.pyplot as plt
```

```python
df_adress = pd.read_csv (url)
```

### Delete the Unnamed 0 column

```python
df_adress
```

```python
df_adress.pop('Unnamed: 0')
```

### Plot the total_bill column histogram

```python
sns.displot(x=df_adress.total_bill)
```

### Create a scatter plot presenting the relationship between total_bill and tip

```python
sns.scatterplot(y=df_adress.total_bill,x=df_adress.tip, data = df_adress);
```

###  Create one image with the relationship of total_bill, tip and size.
#### Hint: It is just one function.

```python
sns.pairplot(df_adress);
```

### Present the relationship between days and total_bill value

```python
sns.boxplot(x=df_adress.day,y=df_adress.total_bill)
```

### Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
sns.stripplot(data = df_adress, y=df_adress.day,x=df_adress.tip,hue=df_adress.sex,jitter= True)
```

###  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
sns.boxplot(x=df_adress.day,y=df_adress.total_bill,hue=df_adress.time)
```

### Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
# creates FacetGrid
g = sns.FacetGrid(df_adress, col = "time")
g.map(plt.hist, "tip");
```

### Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
x= sns.FacetGrid(df_adress,col='sex',hue ="smoker")
x.map(plt.scatter,"total_bill","tip",alpha =.7)
x.add_legend;
```
